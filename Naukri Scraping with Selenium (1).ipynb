Naukri Scraping using Selenium 

```python
from selenium.webdriver.common.by import By
```

```python
import pandas as pd
from selenium import webdriver
from datetime import datetime, timedelta
import time

scraped_detailss= []

for i in range(1,2):
    url="https://www.naukri.com/data-science-jobs"
    driver = webdriver.Chrome('C:/Users/Ritwaj.Dubey/Downloads/chromedriver_win32/chromedriver.exe')
    driver.get(url+"-"+str(i))
    
    time.sleep(5)
    
    
    job_links=[my_elem.get_attribute("href") for my_elem in driver.find_elements(By.XPATH,"//div[@class='info fleft']/a")]
    for links in job_links:
        driver.get(links)
        
        time.sleep(3)
        job_title=driver.find_elements(By.XPATH,"//h1[@class='jd-header-title']")
        company_name=driver.find_elements(By.XPATH,"//a[@class='pad-rt-8']")
        exp_years=driver.find_elements(By.XPATH,"//div[@class='exp']")
        salary_details=driver.find_elements(By.XPATH,"//div[@class='salary']")
        posted_days=driver.find_elements(By.XPATH,"//div[@class='jd-stats']")
        descp_details=driver.find_elements(By.XPATH,"//div[@class='dang-inner-html']")
        job_details=driver.find_elements(By.XPATH,"//div[@class='other-details']")
        edu_details=driver.find_elements(By.XPATH,"//div[@class='education']")
        key_skills=driver.find_elements(By.XPATH,"//div[@class='key-skill']")
        
        for job,company,exp,salary,posted,descp,details,edu,skills in zip(job_title,company_name,exp_years,salary_details,posted_days,descp_details,job_details,edu_details,key_skills):
            scraped_detailss.append({"Job":job.text,"Company":company.text,"Exp_years":exp.text,"Salary":salary.text,"Posted":posted.text,"Description":descp.text,"Job Details":details.text,"Education Details":edu.text,"Key Skills":skills.text,"Job Links":links})
            naukri_df=pd.DataFrame(scraped_detailss)
```
